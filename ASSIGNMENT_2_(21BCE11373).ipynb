```python
import pandas as pd

# Load the dataset from the Excel file
df = pd.read_excel('House Price India.xlsx')  

df.head()
```

# task 3

# uni variate visualization

```python
import matplotlib.pyplot as plt

#histogram

plt.hist(df['Built Year'], bins=10)  
plt.xlabel('year ')
plt.ylabel('no. of houses ')
plt.title('Histogram of Built Year')
plt.show()
```

```python
# Scatter plot

plt.scatter(df['living area'], df['lot area'])
plt.xlabel('living area')
plt.ylabel('lot area')
plt.title('Scatter Plot of living area vs. lot area')
plt.show()
```

# multi variate analysis

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = df['number of bedrooms']
y = df['number of bathrooms']
z = df['number of floors']

ax.scatter(x, y, z, c='b', marker='o')

ax.set_xlabel('number of bedrooms')
ax.set_ylabel('number of bathrooms')
ax.set_zlabel('number of floors')

plt.title('3D Scatter Plot')
plt.show()
```
